# Exploratory Data Analysis (EDA)

```
import pandas as pd
import numpy as np
import sweetviz as sv
import matplotlib.pyplot as plt
import seaborn as sns
```

```
# load the data
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

The target distribution shows major class imbalance which will be a key challenge for modeling.

```
# check out the class imbalance for the target variable 
print(train['Irrigation_Need'].value_counts(normalize=True))

# visualize it
plt.figure(figsize=(6,4))
sns.countplot(data=train, x='Irrigation_Need')
plt.title('Distribution of Irrigation Need')
plt.show()
```

```
# generate eda report with sweetviz
try:
    report = sv.analyze(train)
    report.show_html('irrigation_eda_report.html', open_browser=False)
except Exception as e:
    print(f"Sweetviz report generation skipped: {e}")
```
